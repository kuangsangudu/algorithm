# 6278. Count the Digits That Divide a Number

```python
class Solution:
    def countDigits(self, num: int) -> int:
        start = num
        ret = 0
        while start:
            if num % (start % 10) == 0:
                ret += 1
            start //= 10
        return ret
```

# 6279. Distinct Prime Factors of Product of Array

```python
class Solution:
    def distinctPrimeFactors(self, nums: List[int]) -> int:
        re = 1
        for n in nums:
            re *= n
        
        ret = 0
        for i in range(2, 1000):
            if re % i == 0:
                ret += 1
                while re % i == 0:
                    re % = i
        return ret
```

```python
class Solution:
    def distinctPrimeFactors(self, nums: List[int]) -> int:
        s = set()
        for x in nums:
            i = 2
            while i * i <= x:
                if x % i == 0:
                    s.add(i)
                    x //= i
                    while x % i == 0:
                        x //= i
                i += 1
            if x > 1:
                s.add(x)
        return len(s)
```

# 6196. Partition String Into Substrings With Values at Most K

```python
class Solution:
    def minimumPartition(self, s: str, k: int) -> int:
        s = list(map(int, list(s)))
        inf = float("inf")
        ret = [inf] * (len(s)+1)
        ret[0] = 0
        for i in range(len(s)):
            sums = 0
            for j in range(i, len(s)):
                sums = sums*10+s[j]
                if  sums<= k:
                    ret[j+1] = min(ret[j+1], ret[i] + 1)
                else:
                    break
        return ret[-1] if ret[-1] != float("inf") else -1
```

# 6280. Closest Prime Numbers in Range

```python
class oula:
    def __init__(self, n):
        self.N = n
        self.isprime = [True] * self.N
        self.isprime[0] = False
        self.isprime[1] = False
        self.prime = []
        self.__init()
        
    def __init(self):
        cnt = 0
        for i in range(2, self.N):
            if self.isprime[i]:
                self.prime.append(i)
                cnt += 1
            for j in range(cnt):
                if i * self.prime[j] >= self.N:
                    break
                self.isprime[i * self.prime[j]] = False
                if i % self.prime[j]== 0:
                    break
                
    
    def judge(self, n):
        return self.isprime[n]
    
    def get_nthprime(self, n):
        return self.prime[n]
    
class Eratosthenes:
    def __init__(self, n):
        self.N = n
        self.isprime = [True] * self.N
        self.isprime[0] = False
        self.isprime[1] = False
        self.prime = []
        self.__init()
        
    def __init(self):
        for i in range(2, self.N):
            if self.isprime[i]:
                self.prime.append(i)
                j = i * 2
                while j < self.N:
                    self.isprime[j] = False
                    j += i
    
    def judge(self, n):
        return self.isprime[n]
    
    def get_nthprime(self, n):
        return self.prime[n]

class Solution:
    def closestPrimes(self, left: int, right: int) -> List[int]:
        nums = oula(right+10).prime
        i = bisect.bisect_left(nums, left)
        j = bisect.bisect_left(nums, right)
        if j >= len(nums):
            j -= 1
        j = j-1 if nums[j] > right else j
        mins = float("inf")
        ret = [-1, -1]
        for idx in range(i, j):
            if idx+1 <= min(len(nums)-1, j) and nums[idx+1]-nums[idx] < mins:
                ret = [nums[idx], nums[idx+1]]
                mins = nums[idx+1]-nums[idx]
        return ret
```
