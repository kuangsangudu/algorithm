# 6224. Number of Subarrays With GCD Equal to K

```python
# o（n2logn）
class Solution:
    def subarrayGCD(self, nums: List[int], k: int) -> int:
        ret = 0
        for i, n in enumerate(nums):
            if n == k:
                ret += 1
            mins = n
            for j in range(i+1, len(nums)):
                mins = math.gcd(mins, nums[j])
                if mins == k:
                    ret += 1
                
        return ret
```

# 6216. Minimum Cost to Make Array Equal

```python
# o(m) m为nums[i]的数据范围
class Solution:
    def minCost(self, nums: List[int], cost: List[int]) -> int:
        maxs = max(nums)
        sk = [0] * (maxs+2)
        sb = [0] * (maxs+2)
        def add(l, r, k, b):
            if l > r:
                return
            sk[l] += k
            sk[r+1] -= k
            sb[l] += b
            sb[r+1] -= b
        
        for i, n in enumerate(nums):
            add(0, n, -cost[i], n*cost[i])
            add(n+1, maxs-1, cost[i], -n*cost[i])
        
        ret = float("inf")
        for i in range(maxs+1):
            if i > 0:
                sk[i] += sk[i-1]
                sb[i] += sb[i-1]
            ret = min(ret, sk[i]*i+sb[i])
        return ret
```

```python
# from itertools import pairwise(py 3.10)
class Solution:
    def minCost(self, nums: List[int], cost: List[int]) -> int:
        a = sorted(zip(nums, cost))
        ans = total = sum((x - a[0][0]) * c for x, c in a)
        sum_cost = sum(cost)
        for (x0, c), (x1, _) in pairwise(a):
            sum_cost -= c * 2
            total -= sum_cost * (x1 - x0)
            ans = min(ans, total)
        return ans
```

```python
# o(nlogm)
class Solution:
    def minCost(self, nums: List[int], cost: List[int]) -> int:
        def f(mid):
            ret = 0
            for i, num in enumerate(nums):
                ret += abs(num-mid) * cost[i]
            return ret
            
        l, r = min(nums), max(nums)
        while l < r:
            mid = l + (r-l)//2
            if f(mid)>f(mid+1):
                l = mid+1
            else:
                r = mid
        return f(l)
```

# 2449. Minimum Number of Operations to Make Arrays Similar

```python
class Solution:
    def makeSimilar(self, nums: List[int], target: List[int]) -> int:
        nums_odd = []
        nums_even = []
        target_odd, target_even = [], []
        for i in range(len(nums)):
            if nums[i] % 2:
                nums_odd.append(nums[i])
            else:
                nums_odd.append(nums[i])
            if target[i] % 2:
                target_odd.append(target[i])
            else:
                target_even.append(target[i])
        
        nums_odd.sort()
        nums_even.sort()
        target_even.sort()
        target_odd.sort()
        ret = 0
        for i in range(len(nums_odd)):
            ret += abs(nums_odd[i]-target_odd[i])
        for i in range(len(nums_even)):
            ret += abs(nums_even[i]-target_even[i])
        return ret
```
